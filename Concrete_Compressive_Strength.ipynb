# **Prediction of Concrete Compressive Strength using Linear Regression**

```python
# Step 1: Import Library
import pandas as pd
```

```python
# Step 2: Import Data
concrete=pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Concrete%20Compressive%20Strength.csv')
```

```python
concrete.head()
```

```python
concrete.columns
```

```python
# Step 3: Define x and y
y=concrete['Concrete Compressive Strength(MPa, megapascals) ']
```

```python
x=concrete[['Cement (kg in a m^3 mixture)',
'Blast Furnace Slag (kg in a m^3 mixture)',
'Fly Ash (kg in a m^3 mixture)', 'Water (kg in a m^3 mixture)',
'Superplasticizer (kg in a m^3 mixture)',
'Coarse Aggregate (kg in a m^3 mixture)',
'Fine Aggregate (kg in a m^3 mixture)', 'Age (day)']]
```

```python
# Step 4: train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
# Step 5: Choose the model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
# Step 6: Train the model
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

Clearly, the equation of best fit here is y=59.45903971993882 + (0.12544516)a + (0.11677346)b + (0.09002765)c + (-0.09095886)d + (0.39490648)e  = (0.02805813)f + (0.0363444)g + (0.11394849)h

where a,b,c,d,e,f,g and h are the parameters i.e. 'Cement (kg in a m^3 mixture)',
'Blast Furnace Slag (kg in a m^3 mixture)',
'Fly Ash (kg in a m^3 mixture)', 'Water (kg in a m^3 mixture)',
'Superplasticizer (kg in a m^3 mixture)',
'Coarse Aggregate (kg in a m^3 mixture)',
'Fine Aggregate (kg in a m^3 mixture)', 'Age (day)'

```python
# Step 7: Predict with the model
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
# Step 8: Accuracy of model
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

This means that error in prediction is 32.19% approx and so our model is 100-32.19=67.81% accurate.
